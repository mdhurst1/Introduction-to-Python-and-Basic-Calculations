## Exercise 1: Stress and Pressure
Stress is a form of pressure resulting from a force being applied on an object. The amount of pressure (P) depends on the force (F) being applied and the surface area (A) over which it is distributed. 

$$
\begin{equation}
P = {{F}\over{A}} \tag{1} \label{eq:1}
\end{equation}
$$

Force is measured in units called Newtons [N], where 1N = 1 kg m s$^{-2}$ (one kilogram metre per second-squared). When a force of 1N is applied to a 1 m$^2$ surface area it results in a pressure of 1 Pascal (Pa). So what are the units of a Pascal? 
- Well **1N/m$^2$** which is the same as saying **1N m$^{-2}$**. 
- Adding the unit definition of a Newton from above (1N = 1 kg m s$^{-2}$) we get **kg m s$^{-2}$ m$^{-2}$**.
- Some of these can cancel since in there we have **m/m$^2$**. So the units for a Pa are **kg m$^{-1}$ s$^{-2}$**.

In atmospheric sciences, atmospheric pressures are often referred to in kPa (thousands of Pascals). In Geology, pressure is often referred to in MPa (millions of Pascals).

### Unit conversion
That last answer gives pressure in bars, and there are a lot of them. We might want to express pressure in different units like Pascals (or kPa or mPa). In order to convert between units we need to know a conversion factor. 1 bar = 100,000 Pa. So to convert to Pa we multiply by the conversion factor.

```python
Pressure_Bars = 9000. # in bars
ConvertPa = 100000. # conversion factor to Pascals
Pressure_Pa = Pressure_Bars * ConvertPa
print("Pressure in Pa", Pressure_Pa, "Pascals")
```
